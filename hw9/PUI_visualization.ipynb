```python
import geopandas as gp
import pandas as pd
import numpy as np
import os
import json
import pylab as pl
import scipy.stats
%pylab inline
```

```python
citi = pd.read_csv('201501-citibike-tripdata 2.csv')
```

```python
def hour(ts):
    return ts.hour
def year(ts):
    return ts.year 
```

```python
citi['time'] = pd.to_datetime(citi['starttime'])
citi['time2'] = pd.to_datetime(citi['starttime'])
```

```python
citi['hour'] = citi['time'].apply(hour)
citi['year'] = citi['time'].apply(year)
```

```python
citi['ageD'] = 2015-citi['birth year'][(citi['usertype'] == 'Subscriber') & 
                                       (citi['hour']>6) & (citi['hour']<18)]
citi['ageN'] = 2015-citi['birth year'][(citi['usertype'] == 'Subscriber') & 
                                       (citi['hour']<6) | (citi['hour']>18)]
```

## Comparing day and night age ranges for Citibike riders 

```python
bin = np.arange(10,99,10)
plt.figure()

figsize(20,4)

fig1 = citi.ageD.groupby(pd.cut(citi.ageD,bin)).agg([count_nonzero]).plot(kind='area', color = 'Blue', alpha = 0.4, fontsize=14)
pl.xlabel('Age Day', fontsize=22)
pl.ylabel('No of riders', fontsize=20)

pl.title('Comparing day and night age ranges for Citibike riders',fontsize=25)

fig2 = citi.ageN.groupby(pd.cut(citi.ageN,bin)).agg([count_nonzero]).plot(kind='area', color = 'Green', alpha = 0.4,fontsize=14)
pl.xlabel('Age Night', fontsize=22)
pl.ylabel('No of riders',fontsize=20 )
plt.show
```
